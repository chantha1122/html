linked List Push operation


```python
class Node:
  def __init__(self,data):
    self.data=data
    self.next=None
```

```python
class LinkedList:
  def __init__(self):
    self.head=None

  #add a node at the beginning (head)
  def push(self,data):
    new_node=Node(data)
    new_node.next=self.head
    self.head=new_node
  #add the node at the end(tail)
  def append(self,data):
    new_node=Node(data)
    if self.head is None:
      self.head=new_node
      return

    last= self.head
    while last.next:
      last=last.next
    last.next=new_node

  #display the linkes list
  def display(self):
    elements=[]
    current=self.head
    while current:
      elements.append(current.data)
      current=current.next
    return elements
```

```python
#example usage
linked_list=LinkedList()
linked_list.push(10)
linked_list.push(20)
linked_list.push(30)
linked_list.append(50)

print(f"Linked list: {linked_list.display()}")
```

```python
class BoundedStack:
  def __init__(self,max_size=10):
    self.items=[] #this is a stack
    self.max_size=max_size

#push an element with stack overflow check
  def push(self,item):
    if self.is_full():
      raise StackOverflowError("Stack is full! cannot push more element.")
    self.items.append(item)

#pop element with underflow check
  def pop(self):
    if self.is_empty():
      raise StackUnderflowError("Stack is empty!")
    return self.items.pop()

#check stack empty
  def is_empty(self):
     return len(self.items)==0

#check stack full
  def is_full(self):
    return len(self.items) >= self.max_size

#check size of the stack
  def size(self):
    return len(self.items)


class StackOverflowError(Exception): #this create class to store exeption error
  pass

class StackUnderflowError(Exception):
  pass
```

Singly Linked list(students)

```python
class Node:
  def __init__(self,student_id,name,score):
    self.student_id=student_id
    self.name=name
    self.score=score
    self.next=None

class SinglyLinkedList:
  def __init__(self):
    self.head=None

  def append(self,student_id,name,score):
    new_node=Node(student_id,name,score)#--> this is object of class Node

    #- Check if the list is empty
    if not self.head: #- If self.head is None, the list has no nodes yet.
      self.head= new_node # - the new node becomes the head (first node)
      return

    #Traverse to the end
    cur=self.head #(cur= first node)
    while cur.next: #Start at the head.
      cur=cur.next  #- Move forward one node at a time using .next
    cur.next=new_node #- Stop when you reach the last node (its .next is None)

  def delete(self,student_id):
    cur=self.head
    if cur and cur.student_id==student_id:
      self.head= cur.next
      return
    prev=None
    while cur and cur.student_id != student_id:
      prev=cur
      cur=cur.next
    if cur:
      prev.next=cur.next

  def print_list(self):
    cur=self.head
    print("student ID | Name   | Score")
    print("="*30)
    while cur:
      print(f"{cur.student_id:<10} | {cur.name:<8} | {cur.score}")
      cur=cur.next
```

```python
#example with real data
students= SinglyLinkedList()
students.append("H!2","yuki",24)
students.append("H#$","ahha",35)
students.append("S03", "Charlie", 78)
students.append("S04", "Diana", 88)

print("===Original Student List ===")
students.print_list()

students.delete("H!2")
print("\n===Student List After Deletion===")
students.print_list()
```

Singly linked list using excel data

```python
# ===============================
# Example: Singly Linked List using Excel Data
# ===============================

import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
#load the excel file
df = pd.read_excel("/content/drive/MyDrive/year2_semester1/ADS/my_data/student_list.xlsx")

# Step 2: Show the first few rows
print("=== Data Preview (head) ===")
print(df.head(), "\n")

# Let's assume your dataset has columns like:
# Student_ID | Name | Score
# If your column names are different, update below
cols = list(df.columns)
print("Detected Columns:", cols, "\n")

# Step 3: Define the Node and Linked List classes
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class SinglyLinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        """Add new node to the end of the list"""
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return


        current = self.head
        while current.next:
            current = current.next
        current.next = new_node

    def print_list(self):
        """Print all nodes in the linked list"""
        current = self.head
        print("=== Singly Linked List Data ===")
        while current:
            print(current.data)
            current = current.next
        print()

# Step 4: Convert Excel data rows into Python dictionaries
records = df.to_dict(orient="records")

# Step 5: Create a Singly Linked List and add each record
linked_list = SinglyLinkedList()
for record in records:
    linked_list.append(record)

# Step 6: Display the linked list
linked_list.print_list()
```
